```python
# exist_data = pd.read_excel('2023 Product Benchmark --29.xlsx', 'Market Share')
# file = pd.read_csv("New Microsoft Excel Worksheet.csv")
file = pd.read_csv("Product Benchmark MS_547.csv", encoding='windows-1252')
col_lst = ['Time Period(s)','Size Cluster','Quantum Dot Technology','TV Series','Brand','Outlet','Price Cluster',
           'Software Service Name','Dollars','Units ']
file.columns = col_lst
file = file.drop(file.index[0:3]).reset_index()
file = file.drop('index', axis = 1)
file = file.dropna()
file
```

```python
DS_list = []
PS_list = []
Weeks_list = []
Brand_list = ['VIZIO','Samsung','Hisense','Sony','LG','TCL']
year_size = 52
cur_year = 2023

size_list = ['32','40','43','50','55-58','65','70','75','77+']
price_list = ['<500$', '500$-700$', '700$-800$','800$-1000$','1000$-1500$','1500$-2000$','2000$+']

Weeks_list = file["Time Period(s)"].unique().tolist()
weeks_lists = [Weeks_list[i : i + year_size] for i in range(0, len(Weeks_list), year_size)]

file["Price Cluster"] = file["Price Cluster"].replace('<200$', '<500$')
file["Price Cluster"] = file["Price Cluster"].replace('200$-300$', '<500$')
file["Price Cluster"] = file["Price Cluster"].replace('300$-500$', '<500$')
file["Price Cluster"] = file["Price Cluster"].replace('2000$-2500$', '2000$+')
file["Price Cluster"] = file["Price Cluster"].replace('2500$+', '2000$+')  
                                                      
file['Size Cluster'] = file['Size Cluster'].replace('55','55-58')
file['Size Cluster'] = file['Size Cluster'].replace('58','55-58')   
file['Size Cluster'] = file['Size Cluster'].replace('>75','77+')

file['Dollars'] = file['Dollars'].astype(str)
file['Dollars'] = file['Dollars'].str.replace('(','-')
file['Dollars'] = file['Dollars'].str.replace(')','')
file['Dollars'] = file['Dollars'].str.replace('$','')
file['Dollars'] = file['Dollars'].str.replace(',','')
file['Dollars'] = file['Dollars'].astype(float)

file['Units '] = file['Units '].astype(str)
file['Units '] = file['Units '].str.replace(',','')
file['Units '] = file['Units '].astype(float)


file
```

```python
def sum_calculation(table_name, col_name):
    return table_name[col_name].sum()

def fix_dataframe(df, size_price, ilist):
    # Set 'Size Cluster' as index, then reindex using size_list
    df = df.set_index(size_price).reindex(ilist)
        
    
    # Fill NaN values (this step is optional and depends on your requirement)
    df.fillna(0, inplace=True)
    
    # Reset the index
    return df.reset_index()

def YTD_claculation(file, file_brand, table_name, brand_name, weeks_lists, size_list, cur_year, ytd_colname,size_price):
    
    if table_name == 'Remaining Retail':
        if size_price == 'Size Cluster':
            ytd_table = file_brand[(file_brand['Time Period(s)'].isin(weeks_lists[-1])) 
                                   & (file_brand[size_price].isin(size_list))
                                   & (file_brand['Outlet'] == 'Remaining Retail')]
        else:
             ytd_table = file_brand[(file_brand['Time Period(s)'].isin(weeks_lists[-1])) 
                                   & (file_brand[size_price].isin(price_list))
                                   & (file_brand['Outlet'] == 'Remaining Retail')]

    else:
        if size_price == 'Size Cluster':
            ytd_table = file_brand[(file_brand['Time Period(s)'].isin(weeks_lists[-1])) 
                                   & (file_brand[size_price].isin(size_list))]
        else:
            ytd_table = file_brand[(file_brand['Time Period(s)'].isin(weeks_lists[-1])) 
                                   & (file_brand[size_price].isin(price_list))]
        
    if brand_name == 'Industry':
        if table_name == 'Remaining Retail':
            temp = file[(file['Time Period(s)'].isin(weeks_lists[-1])) & (file['Outlet'] == 'Remaining Retail')]
        else:
            temp = file[file['Time Period(s)'].isin(weeks_lists[-1])]
        ytd_sum = sum_calculation(temp, ytd_colname)
    else:
        if size_price == 'Size Cluster' and table_name == 'Remaining Retail':
            temp = file[(file['Time Period(s)'].isin(weeks_lists[-1]))
                        & (file[size_price].isin(size_list)) & (file['Outlet'] == 'Remaining Retail')]
        elif size_price == 'Size Cluster':
            temp = file[(file['Time Period(s)'].isin(weeks_lists[-1]))
                        & (file[size_price].isin(size_list))]
        else:
            temp = file[(file['Time Period(s)'].isin(weeks_lists[-1]))
                        & (file[size_price].isin(price_list))]
            
            
        ytd_sum = temp.groupby(size_price)[ytd_colname].sum().reset_index()
        if size_price == 'Size Cluster':
            ytd_sum = fix_dataframe(ytd_sum, size_price, size_list)
        else:
            ytd_sum = fix_dataframe(ytd_sum, size_price, price_list)
        ytd_sum = ytd_sum[ytd_colname]
        
        
    ds_ytd_table = ytd_table.groupby(size_price)[ytd_colname].sum().reset_index()
    
    if size_price == 'Size Cluster':
        ds_ytd_table = fix_dataframe(ds_ytd_table, size_price, size_list)
    else:
        ds_ytd_table = fix_dataframe(ds_ytd_table, size_price, price_list)
        
        
        
    if table_name == 'Unit number':
        ds_ytd_table[f'{cur_year}YTD'] = ds_ytd_table[ytd_colname] 
    else:
        
        if size_price == 'Price Cluster':
            ds_ytd_table[f'{cur_year}YTD'] = ds_ytd_table[ytd_colname] / ytd_sum * 100
            lst = ['500$+', 0]
            if brand_name == 'Industry':
                file_500pls_sum = file[(file['Time Period(s)'].isin(weeks_lists[-1]))
                                       & (file[size_price].isin(['500$-700$', '700$-800$','800$-1000$',
                                                                 '1000$-1500$','1500$-2000$','2000$+']))].sum()
                file_500pls_sum = file_500pls_sum['Dollars']
                
                
                file_500pls_sum_overall = file[(file['Time Period(s)'].isin(weeks_lists[-1]))
                                           & (file[size_price].isin(price_list))].sum()
                file_500pls_sum_overall = file_500pls_sum_overall['Dollars']
                
            else:
                file_500pls_sum = file_brand[(file_brand['Time Period(s)'].isin(weeks_lists[-1]))
                                             & (file_brand[size_price].isin(['500$-700$', '700$-800$','800$-1000$',
                                                                             '1000$-1500$','1500$-2000$','2000$+']))].sum()
                file_500pls_sum = file_500pls_sum['Dollars']
                
                file_500pls_sum_overall = file[(file['Time Period(s)'].isin(weeks_lists[-1]))
                                               & (file[size_price].isin(['500$-700$', '700$-800$','800$-1000$',
                                                                         '1000$-1500$','1500$-2000$','2000$+']))].sum()
                file_500pls_sum_overall = file_500pls_sum_overall['Dollars']
                
            result_500pls = file_500pls_sum / file_500pls_sum_overall * 100
            lst.append(result_500pls)
            lst_df = pd.DataFrame([lst], columns=ds_ytd_table.columns)
            ds_ytd_table = pd.concat([lst_df, ds_ytd_table], ignore_index=True)
        else:
            ds_ytd_table[f'{cur_year}YTD'] = ds_ytd_table[ytd_colname] / ytd_sum * 100

    ds_ytd_table = ds_ytd_table.drop(ytd_colname, axis=1)
    
    if table_name == 'Remaining Retail':
        file_total = file[(file['Time Period(s)'].isin(weeks_lists[-1]))
                         & (file['Outlet'] == 'Remaining Retail')]
        file_brand_total = file_brand[(file_brand['Time Period(s)'].isin(weeks_lists[-1])
                                      & (file_brand['Outlet'] == 'Remaining Retail'))]
    else:
        file_total = file[file['Time Period(s)'].isin(weeks_lists[-1])]
        file_brand_total = file_brand[file_brand['Time Period(s)'].isin(weeks_lists[-1])]
    
    brand_total = sum_calculation(file_brand_total, ytd_colname)
    grand_total = sum_calculation(file_total, ytd_colname)
    
    total = brand_total / grand_total * 100
        
    return ds_ytd_table, total


def ratio_calculation(file, file_brand, table_name, brand_name, weeks_lists, size_list, week_bf_curwk, cur_year, ytd_colname, size_price):
    #被除数
    if table_name == 'Remaining Retail':
        if size_price == 'Size Cluster':
            ds_in_table = file_brand[(file_brand['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]) 
                                     & (file_brand[size_price].isin(size_list))
                                     & (file_brand['Outlet'] == 'Remaining Retail')]
        else:
            ds_in_table = file_brand[(file_brand['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]) 
                                     & (file_brand[size_price].isin(price_list))
                                     & (file_brand['Outlet'] == 'Remaining Retail')]
            
            
    else:
        if size_price == 'Size Cluster':
            ds_in_table = file_brand[(file_brand['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]) 
                                     & (file_brand[size_price].isin(size_list))]
        else:
            ds_in_table = file_brand[(file_brand['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]) 
                                     & (file_brand[size_price].isin(price_list))]
        
    #底
    if brand_name == 'Industry':
        if table_name == 'Remaining Retail':
            temp1 = file[(file['Time Period(s)'] == weeks_lists[-1][week_bf_curwk])
                        & (file['Outlet'] == 'Remaining Retail')]
        else:
            temp1 = file[file['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]]
        ds_in_table_sum = sum_calculation(temp1, ytd_colname)
    #brand
    else:
        if size_price == 'Size Cluster' and table_name == 'Remaining Retail':
            temp1 = file[(file['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]) 
                         & (file[size_price].isin(size_list))
                         & (file['Outlet'] == 'Remaining Retail')]
              
            
        elif size_price == 'Size Cluster':
            temp1 = file[(file['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]) 
                         & (file[size_price].isin(size_list))]
             
            
        else:
#             temp1 = file_brand[(file_brand['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]) 
#                          & (file_brand[size_price].isin(price_list))]
            temp1 = file[(file['Time Period(s)'] == weeks_lists[-1][week_bf_curwk])
                         & (file[size_price].isin(price_list))]

            
        ds_in_table_sum = temp1.groupby(size_price)[ytd_colname].sum().reset_index()
        if size_price == 'Size Cluster':
            ds_in_table_sum = fix_dataframe(ds_in_table_sum, size_price, size_list)
        else:
            ds_in_table_sum = fix_dataframe(ds_in_table_sum, size_price, price_list)
        ds_in_table_sum = ds_in_table_sum[ytd_colname]  
        
    ds_in_table_gb = ds_in_table.groupby(size_price)[ytd_colname].sum().reset_index()
    
    if size_price == 'Size Cluster':
        ds_in_table_gb = fix_dataframe(ds_in_table_gb, size_price, size_list)
    else:
        ds_in_table_gb = fix_dataframe(ds_in_table_gb, size_price, price_list)
        
        
    if table_name == 'Unit number':
        ds_in_table_gb[f'{cur_year - 2000}W{len(weeks_lists[-1])+week_bf_curwk+1}'] = ds_in_table_gb[ytd_colname]
        
    else:
        if size_price == 'Price Cluster':
            ds_in_table_gb[f'{cur_year - 2000}W{len(weeks_lists[-1])+week_bf_curwk+1}'] = (ds_in_table_gb[ytd_colname] / ds_in_table_sum) * 100
            lst = ['500$+', 0]
            if brand_name == 'Industry':
                file_500pls_sum = file[(file['Time Period(s)'] == weeks_lists[-1][week_bf_curwk])
                                       & (file[size_price].isin(['500$-700$', '700$-800$','800$-1000$',
                                                                 '1000$-1500$','1500$-2000$','2000$+']))].sum()
                file_500pls_sum = file_500pls_sum['Dollars']
                
                file_500pls_sum_overall = file[(file['Time Period(s)'] == weeks_lists[-1][week_bf_curwk])
                                               & (file[size_price].isin(price_list))].sum()
                file_500pls_sum_overall = file_500pls_sum_overall['Dollars']
                
            else:
                file_500pls_sum = file_brand[(file_brand['Time Period(s)'] == weeks_lists[-1][week_bf_curwk])
                                             & (file_brand[size_price].isin(['500$-700$', '700$-800$','800$-1000$',
                                                                             '1000$-1500$','1500$-2000$','2000$+']))].sum()
                file_500pls_sum = file_500pls_sum['Dollars']
                file_500pls_sum_overall = file[(file['Time Period(s)'] == weeks_lists[-1][week_bf_curwk])
                                               & (file[size_price].isin(['500$-700$', '700$-800$','800$-1000$',
                                                                         '1000$-1500$','1500$-2000$','2000$+']))].sum()
                file_500pls_sum_overall = file_500pls_sum_overall['Dollars']
                
            result_500pls = file_500pls_sum / file_500pls_sum_overall * 100
            lst.append(result_500pls)
            lst_df = pd.DataFrame([lst], columns=ds_in_table_gb.columns)
            ds_in_table_gb = pd.concat([lst_df, ds_in_table_gb], ignore_index=True)
        else:
            ds_in_table_gb[f'{cur_year - 2000}W{len(weeks_lists[-1])+week_bf_curwk+1}'] = (ds_in_table_gb[ytd_colname] / ds_in_table_sum) * 100
        
    ds_in_table_gb = ds_in_table_gb.drop(ytd_colname, axis=1)
    
    if table_name == 'Remaining Retail':
        file_total = file[(file['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]) 
                          & (file['Outlet'] == 'Remaining Retail')]
        file_brand_total = file_brand[(file_brand['Time Period(s)'] == weeks_lists[-1][week_bf_curwk])
                                      & (file_brand['Outlet'] == 'Remaining Retail')]
        
    else:
        
        file_total = file[file['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]]
        file_brand_total = file_brand[file_brand['Time Period(s)'] == weeks_lists[-1][week_bf_curwk]]
    

    brand_total = sum_calculation(file_brand_total, ytd_colname)
    grand_total = sum_calculation(file_total, ytd_colname)
    
    total = brand_total / grand_total * 100
    
    return ds_in_table_gb, total

def col_merger(data, brand_name, table_name, weeks_lists, size_list, ytd_colname, size_price):
    
    if brand_name != 'Industry':
        data_brand = data[data['Brand'] == brand_name]
    else:
        data_brand = data
    
    in_3week_bf, total1 = ratio_calculation(data, data_brand, table_name, brand_name, weeks_lists, size_list, -4, 2023, ytd_colname, size_price)
    in_2week_bf, total2 = ratio_calculation(data, data_brand, table_name, brand_name, weeks_lists, size_list, -3, 2023, ytd_colname, size_price)
    in_1week_bf, total3 = ratio_calculation(data, data_brand, table_name, brand_name, weeks_lists, size_list, -2, 2023, ytd_colname, size_price)
    in_cur_week, total4 = ratio_calculation(data, data_brand, table_name, brand_name, weeks_lists, size_list, -1, 2023, ytd_colname, size_price)
    in_YTD_week, total0 = YTD_claculation(data, data_brand, table_name, brand_name, weeks_lists, size_list, cur_year, ytd_colname, size_price)

    #how to merge
    merged_df = in_YTD_week.merge(in_3week_bf, on=size_price).merge(in_2week_bf, on=size_price).merge(in_1week_bf, on=size_price).merge(in_cur_week, on=size_price)
    merged_df = merged_df.sort_index(ascending=False).reset_index()
    merged_df = merged_df.drop('index', axis=1)
    total_list = [round(total0, 2), round(total1, 2), round(total2, 2), round(total3, 2), round(total4, 2)]
    return merged_df.round(2), total_list


def df_connnect(file, table_name, weeks_lists, size_list, ytd_colname, size_price):
    total_lists = ['Total']                              
    merged_df1, total_list0= col_merger(file, 'Industry', table_name, weeks_lists, size_list, ytd_colname, size_price)
    [total_lists.append(i) for i in total_list0]
    
    merged_df2, total_list1= col_merger(file, 'Hisense', table_name, weeks_lists, size_list, ytd_colname, size_price)
    merged_df2 = merged_df2.drop(size_price, axis=1)
    [total_lists.append(i) for i in total_list1]
    
    
    merged_df3, total_list2= col_merger(file, 'TCL', table_name, weeks_lists, size_list, ytd_colname, size_price)
    merged_df3 = merged_df3.drop(size_price, axis=1)
    [total_lists.append(i) for i in total_list2]

    
    merged_df4, total_list3= col_merger(file, 'VIZIO', table_name, weeks_lists, size_list, ytd_colname, size_price)
    merged_df4 = merged_df4.drop(size_price, axis=1)
    [total_lists.append(i) for i in total_list3]
    
    merged_df5, total_list4= col_merger(file, 'Samsung', table_name, weeks_lists, size_list, ytd_colname, size_price)
    merged_df5 = merged_df5.drop(size_price, axis=1)
    [total_lists.append(i) for i in total_list4]
    
    merged_df6, total_list5 = col_merger(file, 'LG', table_name, weeks_lists, size_list, ytd_colname, size_price)
    merged_df6 = merged_df6.drop(size_price, axis=1)
    [total_lists.append(i) for i in total_list5]
    
    merged_df7, total_list6 = col_merger(file, 'Sony', table_name, weeks_lists, size_list, ytd_colname, size_price)
    merged_df7 = merged_df7.drop(size_price, axis=1)
    [total_lists.append(i) for i in total_list6]
    
    return  pd.concat([merged_df1, merged_df2, merged_df3, merged_df4, merged_df5, 
                           merged_df6, merged_df7], axis=1), total_lists

goodday_ds, total_lists = df_connnect(file, None, weeks_lists, size_list, 'Dollars','Size Cluster')
print(total_lists)
# goodday_ds_RR = goodday_ds
goodday_ds_finale = pd.DataFrame([total_lists], columns=goodday_ds.columns)
goodday_ds = pd.concat([goodday_ds_finale, goodday_ds], ignore_index=True)

goodday_ds_temp = goodday_ds.iloc[:,1:]
goodday_ds_temp[goodday_ds_temp < 0] = 0.0
goodday_ds_temp.insert(0, 'Size Cluster', goodday_ds['Size Cluster'])


goodday_ds_final = goodday_ds_temp.applymap(lambda x: str(x) + '%')
goodday_ds_final = goodday_ds_final.drop('Size Cluster', axis=1)
goodday_ds_final.insert(0, 'Size Cluster', goodday_ds_temp['Size Cluster'])
goodday_ds_final
```

```python
def multiple_dfs(df_list, sheets, file_name, spaces):
    writer = pd.ExcelWriter(file_name,engine='xlsxwriter')   
    row = 0
    for dataframe in df_list:
        dataframe.to_excel(writer,sheet_name=sheets,startrow=row , startcol=0)   
        row = row + len(dataframe.index) + spaces + 1
    writer.save()

date = datetime.today().strftime('%Y-%m-%d')
dfs = [goodday_ds_final,goodday_us_final, goodday_RR_final, goodday_BPI,goodday_PB_final,goodday_UN,goodday_ASP]

multiple_dfs(dfs, 'Sheet1', f'Market Share {date}.xlsx', 3)
```
