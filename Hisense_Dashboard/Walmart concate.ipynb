```python
import pandas as pd
df = pd.read_excel('Google-TV.xlsx')
df.columns = df.loc[0]
df = df.drop(df.index[0])
df = df.reset_index()

col_list = list(df.columns.values)
df
```

```python
df = df.replace('walmart.com','Walmart')
df_updated = df.groupby(['Time Period(s)','Outlet'], as_index=False).agg({
            'index':'mean', 'Units Total':'sum', 'Brand':'first', 'Software Service Name' :'first'

})

df_updated = df_updated.sort_values('index')
df_updated = df_updated[col_list]
df_updated = df_updated.drop('index', axis = 1)
df_updated = df_updated.reset_index()
df_updated = df_updated.drop('index', axis = 1)

df_updated.to_excel('Google-TV-New.xlsx')
df_updated
```
