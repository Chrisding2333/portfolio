```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Result.csv")
df.head()
```

```python
df['normalized_RMSE'] = (df['RMSE'] - df['RMSE'].min()) / (df['RMSE'].max() - df['RMSE'].min())
```

```python
df = df[df["Location_name"] == "Non_London"]
df = df[(df["Dataset_size"] == "25%") | (df["Dataset_size"] == "50%")]
grouped = df.groupby('Model')

plt.figure(figsize=(10, 6))
for name, group in grouped:
    group['RMSE'] = group['RMSE'].mean()
    plt.plot(group['Label'], group['RMSE'], marker='o', label=name)

plt.xlabel('Label')
plt.ylabel('normalized_RMSE')
plt.title('normalized_RMSE vs Label with Model lines')
plt.ylim(47000, 52000)
plt.legend()
plt.show()
```

```python
first_groups = df.groupby('Model')

subgroup_dfs = []

for first_name, first_group in first_groups:
    second_groups = first_group.groupby('Location_name')
    for second_name, subgroup in second_groups:
        subgroup_df = subgroup.copy()
        subgroup_dfs.append(subgroup_df)  # Or you can use a dictionary with a tuple key (first_name, second_name)


```

```python
df1 = subgroup_dfs[0]
df1['normalized_RMSE'] = (df1['RMSE'] - df1['RMSE'].min()) / (df1['RMSE'].max() - df1['RMSE'].min())
df1_name_list = df1['Model'].to_list()
df1_city_list = df1['Location_name'].to_list()

groups = df1.groupby('Label')
for name, group in groups:
    plt.plot(group['Dataset_size'], group['normalized_RMSE'], label=name)

plt.xlabel('Dataset Size')
plt.ylabel('RMSE')
plt.title(f"Data Size vs RMSE on {df1_name_list[0]} for {df1_city_list[0]} dataset")
plt.ylim(-1, 2)
plt.legend()
plt.show()
```

```python
df1 = subgroup_dfs[0]
df1_name_list = df1['Model'].to_list()
df1_city_list = df1['Location_name'].to_list()

groups = df1.groupby('Label')
for name, group in groups:
    plt.plot(group['Dataset_size'], group['Duration'], label=name)

plt.xlabel('Dataset Size')
plt.ylabel('Duration')
plt.title(f"Data Size vs Duration {df1_name_list[0]} for {df1_city_list[0]} dataset")
plt.ylim(-100, 500)
plt.legend()
plt.show()
```
